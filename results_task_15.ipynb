```python
# при помощи NumPy для заданного k сгенерировать матрицу A,
# у которой все элементы нулевые, кроме главной диагонали, 
# побочной диагонали, k-1 главных и побочных над- и поддиагоналей
```

```python
import numpy as np
def special_matrix(k):
    A = np.zeros((k, k))
    A += np.eye(k)
    A += np.eye(k, k=1)
    A += np.eye(k, k=-1)
    B = np.zeros_like(A)
    for i in range(k):
        for j in range(k):
            B[i,k-j-1] = A[i,j]
    A = A+B
    
    return A


print(special_matrix(20))
```

```python
# при помощи NumPy для заданного k сгенерировать матрицу A,
# у которой все элементы нулевые, кроме главной диагонали, 
# побочной диагонали, k-1 главных и побочных над- и поддиагоналей
```

```python
import numpy as np

def func_generating(n,k):
    m = np.zeros((n,n))
    np.fill_diagonal(m, 1)
    np.fill_diagonal(np.fliplr(m), np.ones(3))
    rows, cols = np.indices((n,n))
    for i in range(k):
        row_vals = np.diag(rows, k=i)
        col_vals = np.diag(cols, k=i)
        m[row_vals, col_vals]=1
        row_vals = np.diag(rows, k=-i)
        col_vals = np.diag(cols, k=-i)
        m[row_vals, col_vals]=1
    return m


print(func_generating(5,2))
```

```python
import numpy as np

def NewMatrix(n, k):
    matrix = np.random.randint(1, 9+1, (n, n)) #Создаём матрицу с произвольными значениями
    for i in range (n):  #Зануляем значения на необходимых диагоналях
        for j in range (n):
            if (abs(i - j) > k):
                if (abs(n - 1 - j - i) > k):
                    matrix[i,j] = 0
    return matrix

print("Введите размер матрицы: ")
n = int(input())
print("Введите значение k: ")
k = int(input())
if k <= 0 or n<=0 or k > n/2:
    print("Попробуйте другие значения")
else:
    print(NewMatrix(n, k))
```
