```python
from math import pi, atan
summa = 0
def angle(coor_1,coor_2,j,summa):
    koef_1 = (coor_2[j]-coor_2[j+1])/(coor_1[j]-coor_1[j+1])
    koef_2 = (coor_2[j+1]-coor_2[j+2])/(coor_1[j+1]-coor_1[j+2])
    if koef_1*koef_2 == -1:
        summa += pi/2
    else:
        aangle = atan(abs((koef_1-koef_2)/(1+koef_1*koef_2)))
        if aangle < 0:
            aangle += pi
        summa += aangle
    if j<min(len(coor_1)-3, len(coor_2)-3):
        summa = angle(coor_1,coor_2, j+1, summa)
    return summa
print(angle(X,Y,0,summa))
```
