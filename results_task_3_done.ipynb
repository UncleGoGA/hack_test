### Чердак Александр СКБ171

Оценка: 0.5

Комментарий: Usage of .format and % supposed

```python
# вместо параметров {mission}, {planet} в строке s подставить их значения из словаря p1
# вместо параметров %d в строке s подставить значения из списка p2

p1 = {'mission':'BepiColombo', 'planet':'Меркурия'}
p2 = [9, 7]

s = 'Началась международная миссия {mission} по изучению самой близкой к Солнцу планеты – {planet}. \
Космический аппарат совершит девять пролётов возле планет и преодолеет в общей сложности около %d миллиардов \
километров в течение %d лет на пути к цели.'
```

```python
p1 = {'mission':'BepiColombo', 'planet':'Меркурия'}
p2 = [9, 7]

keylist = list()
itemlist = list()

def concatNewStr(chlist):
    tempstr = str()
    for x in chlist:
        tempstr += str(x)
    return tempstr

def replacePercentD(pstr, pp2):
    charlist = list()
    inc = 0
    for x in pstr:
        charlist.append(x)
    while pstr.find('%d') != -1:
        ind = pstr.find('%d')
        charlist[ind] = pp2[inc]
        charlist[ind + 1] = ' '
        inc += 1
        pstr = concatNewStr(charlist)
    return pstr

for x in p1:
    keylist.append("{" + x + "}")
    itemlist.append(p1[x])

s = 'Началась международная миссия {mission} по изучению самой близкой к Солнцу планеты – {planet}. \
Космический аппарат совершит девять пролётов возле планет и преодолеет в общей сложности около %d миллиардов \
километров в течение %d лет на пути к цели.'

for i in range(len(keylist)):
    s = s.replace(keylist[i],itemlist[i])

print(replacePercentD(s, p2))
```

```python
p1 = {'mission':'BepiColombo', 'planet':'Меркурия'}
p2 = [9, 7]

s = 'Началась международная миссия {mission} по изучению самой близкой к Солнцу планеты – {planet}. \
Космический аппарат совершит девять пролётов возле планет и преодолеет в общей сложности около %d миллиардов \
километров в течение %d лет на пути к цели.'

s1=s.split()
p1k=list(p1.keys())
p1v=list(p1.values())
i=0
reg=0
reg1=0
while i<len(s1):
    if(s1[i]=='%d')&(reg<len(p2)):
        s1[i]=s1[i].replace(s1[i], str(p2[reg]))
        reg+=1
        
    if(s1[i]=='{'+p1k[reg1]+'}')|(s1[i]=='{'+p1k[reg1]+'}'+'.'):
        #print(s1[i])
        s1[i]=s1[i].replace(s1[i], p1v[reg1])
        if(reg1<len(p1v)-1):
            reg1+=1
    
    i+=1
s=''
i=0
while i<len(s1):
    s+=s1[i]
    s+=' '
    i+=1
print(s)
```

```python
# вместо параметров {mission}, {planet} в строке s подставить их значения из словаря p1
# вместо параметров %d в строке s подставить значения из списка p2

p1 = {'mission':'BepiColombo', 'planet':'Меркурия'}
p2 = [9, 7]

s = 'Началась международная миссия {mission} по изучению самой близкой к Солнцу планеты – {planet}. \
Космический аппарат совершит девять пролётов возле планет и преодолеет в общей сложности около %d миллиардов \
километров в течение %d лет на пути к цели.'
```

```python
s = 'Началась международная миссия {mission} по изучению самой близкой к Солнцу планеты – {planet}. \
Космический аппарат совершит девять пролётов возле планет и преодолеет в общей сложности около %d миллиардов \
километров в течение %d лет на пути к цели.'
p1 = {'mission':'BepiColombo', 'planet':'Меркурия'}
p2 = [9, 7]

q = 2 #кол- во параметров

a = list(map(str, p2))

for i in range(0, len(p2)):
    s = s.replace('%d', a[i], 1)

for i in range(0, q):
    a = []
    b = s.find('{')
    e = s.find('}')
    for g in range(b, e + 1):
        a.append(s[g])
    par = ''.join(a)
    length = len(a)
    del a[0]
    del a[length-2]
    par1 = ''.join(a)
    key_exists = par1 in p1
    if key_exists == True:
        s = s.replace(par, p1[par1])
print(s)

#s.format(**p1).replace('%d', '{}').format(*p2) еще одно решение
```
