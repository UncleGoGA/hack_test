```python
def getSentences(text):
    last = 0;
    res = []
    while last != len(text) - 1:
        d = text.find('.', last + 1)
        q = text.find('?', last + 1)
        if (d < q or q == -1):
            res.append(text[last + 1: d + 1].strip().replace('\\', ' ').replace('\n', ' ').replace('  ', ' '))
            last = d
        elif q != -1:
            res.append(text[last + 1: q + 1].strip().replace('\\', ' ').replace('\n', ' ').replace('  ', ' '))
            last = q
    return res
```

```python
def analyze(p):
    z = p.count(',')
    c = p.replace(',',' ').replace(';', '').replace('-', '').replace(':', '').replace('\"', '').replace('  ', ' ').replace('  ', ' ').split(' ')
    x = len(c)
    y = len(list(filter(lambda x: x[0].isupper(), c)))
    return (p, x, y, z)
```

```python
#tup = (p, x, y, z)
def text(s):
    lst = []
    sym = ['?', '!', '...']
    for i in sym:
        s = s.replace(i, '.')
    s = s.replace('\n', ' ')
    s1 = s.split('.')
    
    
    for j in s1:
        p = j.strip() # p - предложение из текста s
        if len(p) != 0:
            j_split = j.split() #Удаляем лишние пробелы в начале каждой строки
            x = len(j_split) # x - количество слов в p
            y = sum(1 for c in j if c.isupper()) # y - количество слов в p, начинающихся на заглавную букву
            z = j.count(',') # z - количество запятых в p

            tup = (p, x, y, z) #Создаём кортеж из неодходимых значений
            lst.append(tup) #Добавляем текущий кортеж в список
        else: #Если строка нулевой длинны, то она не считается
            break

    print(*lst, sep = '\n')

text(s)
```
