```python
def my_dict(s, al):
    s = s.lower().replace('"', '').replace('-','')
    s = s.split()
    d = dict()
    
    # заполнение словаря начальными
    for i in al:
        d[i] = [0, 0]
        
    # первая буква в слове
    for word in s:
        d[word[0]][0] += 1
        # буквы, которых нет в слове
        for i in al:
            if i not in word:
                d[i][1] += 1
    for j in d:
           d[j] = tuple(d[j])
    return d

al = 'ёйцукенгшщзхфывапролджэячсмитбю'
print(my_dict(s, al))
```

```python
import re
import numpy as np

def getRusAlphabet():
    alp = np.empty(32, dtype=object)
    a = ord('а')
    k = 0
    for i in range(a, a + 32):
        alp[k] = chr(i)
        k = k + 1
    
    return alp


def deleteNonLetters(s):
    nonLetters = [',', '.', '-', '?', '"', '!']
    
    for char in nonLetters:
        s = s.replace(char, '')
                
    return s

s = deleteNonLetters(s)
words = s.lower().split()

alp = getRusAlphabet()
starts = np.zeros(len(alp), dtype = int)
notContains = np.zeros(len(alp), dtype = int)
voc = {}


for word in words:
    for i in range(len(alp)):
        if (word[0] == alp[i]):
            starts[i] = starts[i] + 1
        if (word.find(alp[i]) == -1):
            notContains[i] = notContains[i] + 1
        
for i in range(len(alp)):
    voc[str(alp[i])] = (starts[i], notContains[i])
    
for t in voc: 
    print(t, voc[t])
```
